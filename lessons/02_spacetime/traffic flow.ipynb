```python
import numpy
import sympy
from sympy import init_printing
init_printing()
import matplotlib.pyplot as plt                 
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16

```

```python
###variable declarations
v_max=80.0
L=11.0
rho_max=250.0
nx=51
dx=L/(nx-1)
dt=0.001
T=6.0/60 #final time in hours
nt=int(T/dt)


### initial conditions
x = numpy.linspace(0,L,nx)
rho0 = numpy.ones(nx)*10

rho0[10:20] = 50.0

rhon=numpy.empty(nx)
rho=rho0.copy()

for n in range(nt):
    rhon=rho.copy()
    F=v_max*rhon*(1-rhon/rho_max)
    rho[1:]=rhon[1:]-dt/dx*(F[1:]-F[0:-1])
    rho[0]=10.0
    
    v=v_max*(1-rho/rho_max)/3.6 #in m/s

```

```python

print min(v_max*(1-rho0/rho_max)/3.6)
print numpy.mean(v)# change T = 3.0/60
print min(v)# change T = 6.0/60
```

```python
from JSAnimation.IPython_display import display_animation
from matplotlib import animation
rho=rho0.copy()
fig = plt.figure(figsize=(8,6))
ax = plt.axes(xlim=(0,12), ylim=(0,130))
line = ax.plot([], [], color='#003366', ls='--', lw=3)[0]
line2 = ax.plot([], [], 'k-', lw=2)[0]
ax.legend(['flow','velocity'])

def traffic(n):
       
    rhon=rho.copy()
    F=v_max*rhon[:]*(1-rhon[:]/rho_max)
    rho[1:]=rhon[1:]-dt/dx*(F[1:]-F[0:-1])
    rho[0]=10.0
    
    v=v_max*(1-rho/rho_max)
    
    line.set_data(x,rho)
    line2.set_data(x,v)
    

animation.FuncAnimation(fig, traffic,
                        frames=nt, interval=100)
```
